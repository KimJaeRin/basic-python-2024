```python
## 타일종류, 마커표시
map = folium.Map(location=[35.173999, 129.196649])
folium.Marker([35.173999, 129.196649], popup = '송정해수욕장').add_to(map) #마커추가
map
```

```python
## 마커 종류 변경
map = folium.Map(location=[35.173999, 129.196649], zoom_start = 18, tiles = 'cartodb positron')
folium.Marker([35.173999, 129.196649], popup = '송정해수욕장',
              icon = folium.Icon(color= 'pink',icon = 'remove-circle')).add_to(map) #마커추가, 마커종류 변경
map

```
